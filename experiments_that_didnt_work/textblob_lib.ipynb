```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('clean_data.csv')
```

```python
x = df.loc[1, 'clean_text']
```

```python
from textblob import TextBlob
```

```python
blob = TextBlob(x)
blob.correct()
```

```python
lemmatized = []
for v, pos in blob.tags:
    if pos not in ['DT', 'IN', 'PRP']:
        lemmatized.append(v.lemmatize(pos))
x_lemmatized = ' '.join(lemmatized)
x_blob = TextBlob(x_lemmatized)
x_sentiment = x_blob.sentiment.polarity
x_subject = x_blob.sentiment.subjectivity
```

```python
lemmatized_text = []
sentiments = []
subjectivity = []
for i, row in df.iterrows():
    text = row['clean_text']
    blob = TextBlob(text)
    blob.correct()
    lemmatized = []
    for v, pos in blob.tags:
        if pos not in ['DT', 'IN', 'RP', 'CC',
                       'FW', 'PRP', 'CD', 'MD', 'WDT']:
            try:
                lemmatized.append(v.lemmatize(pos))
            except Exception:
                lemmatized.append(v)
    x_lemmatized = ' '.join(lemmatized)
    x_blob = TextBlob(x_lemmatized)
    x_sentiment = x_blob.sentiment.polarity
    x_subject = x_blob.sentiment.subjectivity
    
    lemmatized_text.append(x_lemmatized)
    sentiments.append(x_sentiment)
    subjectivity.append(x_subject)
    print('\r', i, end='')
```

```python
df['text_lem_cor'] = pd.Series(lemmatized_text)
```

```python
df['sentiment'] = pd.Series(sentiments)
df['subjectivity'] = pd.Series(subjectivity)
```

```python
df = df.loc[:, ['text', 'clean_text', 'clean_text_lemma', 'clean_text_neg', 'text_lem_cor', 'sentiment', 'subjectivity', 'class']]
```

```python
df.to_csv('enh_data.csv', index=False)
```
